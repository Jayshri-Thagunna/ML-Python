```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1 Create Dataset
data = {
    "Hours": [1,2,3,4,5,6,7,8],
    "Marks": [20,30,35,50,55,60,70,80],
    "Result": [0,0,0,0,1,1,1,1]   # 0 = Fail , 1 = Pass
}

# 2 Create DataFrame
df = pd.DataFrame(data)

# Features and Target
X = df[["Hours","Marks"]]
y = df["Result"]

# 3 Split Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


# KNN Model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

pred1 = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, pred1))


# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)

pred2 = lr.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred2))


# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)

pred3 = rf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, pred3))
```
