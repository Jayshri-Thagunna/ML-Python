```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'Distance':[10,20,30,40,50],
    'FuelUsed': [1,2,3,4,5]
}

df = pd.DataFrame(data)
x = df[['Distance']]
y = df[['FuelUsed']]

model = LinearRegression()
model.fit(x,y)

yoe =[[25]]
predicted_salary = model.predict(yoe)

print("Predicted salary for 25 years of experience",predicted_salary[0])

plt.scatter(df['Distance'], df['FuelUsed'])
plt.plot(df['Distance'], model.predict(x), color='red')
plt.xlabel('Distance')
plt.ylabel('FuelUsed')
plt.title('Linear Regression')
plt.show()
```
