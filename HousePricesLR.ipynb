```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("datasets/house_prices.csv")
```

```python
print(dataframe)
```

```python
import numpy as np

dataframe_clean = dataframe.dropna(subset=['Carpet Area Numeric', 'Price (in rupees)'])

x = dataframe_clean['Carpet Area Numeric'].values.reshape(-1, 1)
y = dataframe_clean['Price (in rupees)']

model = LinearRegression()
df = pd.DataFrame()


x = dataframe_clean['Carpet Area Numeric']
y = dataframe_clean['Price (in rupees)']


x = x.values.reshape(-1, 1)

model = LinearRegression()
model.fit(x, y)

new_area = np.array([[2000]])
predicted_price = model.predict(new_area)

predicted_line = model.predict(x)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, alpha=0.5, label='Actual Data')
plt.plot(x, predicted_line, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('Carpet Area (sqft)')
plt.ylabel('Amount (Rupees)')
plt.title('House Prices vs Carpet Area - Linear Regression')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
